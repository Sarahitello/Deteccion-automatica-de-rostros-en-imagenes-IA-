
### **2. Importaci√≥n de librer√≠as**
Utilic√© **OpenCV** para el procesamiento de im√°genes y **Matplotlib** para visualizar los resultados.

```python
import cv2
import matplotlib.pyplot as plt
```

**OpenCV** es una de las librer√≠as m√°s utilizadas en visi√≥n por computadora, y el clasificador **Haar Cascade** permite detectar patrones faciales mediante un modelo entrenado con caracter√≠sticas geom√©tricas.

### **3. Descarga del clasificador Haar Cascade**
El clasificador **Haar Cascade** para detecci√≥n frontal de rostros se descarga directamente desde el repositorio oficial de OpenCV.

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```


### **4. Carga del clasificador**
El clasificador se carga en memoria para ser utilizado en la detecci√≥n de rostros.


```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

### **5. Carga de la imagen**
El usuario puede cargar una imagen desde su dispositivo para analizarla.



```python
from google.colab import files
uploaded = files.upload()
```



### **6. Procesamiento de la imagen**
La imagen se convierte a escala de grises para facilitar el an√°lisis, y se aplica el m√©todo **`detectMultiScale`** para identificar regiones que contienen rostros.

```python
image_name = list(uploaded.keys())[0]

img = cv2.imread(image_name)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
```

### **7. Detecci√≥n y visualizaci√≥n de rostros**
Cada rostro detectado se marca con un rect√°ngulo azul que indica su ubicaci√≥n en la imagen.

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

Cada rostro detectado se marca mediante un rect√°ngulo que indica su localizaci√≥n dentro de la imagen.
